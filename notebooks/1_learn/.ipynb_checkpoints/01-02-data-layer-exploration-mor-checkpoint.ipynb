```python
# 1. í™˜ê²½ ì„¤ì • ë° ê¸°ë³¸ ì •ë³´
from pyspark.sql import SparkSession
import os
import glob

# Spark ì—°ê²° (ì´ë¯¸ ì„¤ì •ë˜ì–´ ìˆë‹¤ê³  ê°€ì •)
# spark = ... (ê¸°ì¡´ ì„¤ì • ì‚¬ìš©)

print("ğŸ“Š í˜„ì¬ í…Œì´ë¸” ìƒíƒœ:")
spark.sql("SELECT COUNT(*) as total_records FROM demo.blog.orders").show()

# ë°ì´í„° ë‚ ì§œ ë²”ìœ„ í™•ì¸
spark.sql("""
    SELECT 
        MIN(order_date) as earliest_date,
        MAX(order_date) as latest_date,
        COUNT(DISTINCT order_date) as unique_dates
    FROM demo.blog.orders
""").show()
```

```python
# 2. ì‹¤ì œ Data Files ìœ„ì¹˜ íƒìƒ‰
warehouse_path = "/home/jovyan/data/warehouse"
table_path = f"{warehouse_path}/blog/orders"

print("ğŸ“ í…Œì´ë¸” ë””ë ‰í† ë¦¬ êµ¬ì¡°:")

def analyze_directory(path, prefix=""):
    items = []
    try:
        for item in sorted(os.listdir(path)):
            if item.startswith('.'):
                continue
            item_path = os.path.join(path, item)
            
            if os.path.isfile(item_path):
                size = os.path.getsize(item_path)
                file_type = "ğŸ“„ Data File" if item.endswith('.parquet') else "ğŸ“„ File"
                items.append(f"{prefix}â”œâ”€â”€ {item} ({size:,} bytes) {file_type}")
            elif os.path.isdir(item_path):
                items.append(f"{prefix}â”œâ”€â”€ ğŸ“ {item}/")
                items.extend(analyze_directory(item_path, prefix + "â”‚   "))
    except:
        pass
    return items

structure = analyze_directory(table_path)
for item in structure:
    print(item)
```

```python
# 3. Parquet íŒŒì¼ë“¤ ìƒì„¸ ë¶„ì„ (ìˆ˜ì • ë²„ì „)
from pyspark.sql.functions import min as spark_min, max as spark_max

data_files = glob.glob(f"{table_path}/data/**/*.parquet", recursive=True)

print(f"ğŸ” ë°œê²¬ëœ Parquet íŒŒì¼ ê°œìˆ˜: {len(data_files)}")
print("=" * 50)

for i, file_path in enumerate(data_files[:5]):  # ì²˜ìŒ 5ê°œë§Œ ë¶„ì„
    file_size = os.path.getsize(file_path)
    relative_path = file_path.replace(table_path, "")
    
    print(f"ğŸ“„ íŒŒì¼ #{i+1}: {relative_path}")
    print(f"   í¬ê¸°: {file_size:,} bytes ({file_size/1024:.1f} KB)")
    
    # Sparkìœ¼ë¡œ íŒŒì¼ ë‚´ìš© ë¶„ì„
    df = spark.read.parquet(file_path)
    record_count = df.count()
    
    print(f"   ë ˆì½”ë“œ ìˆ˜: {record_count:,}ê°œ")
    
    # ë‚ ì§œ ë²”ìœ„ í™•ì¸ (ìˆ˜ì •ëœ ë°©ì‹)
    if record_count > 0:
        date_stats = df.agg(
            spark_min("order_date").alias("min_date"),
            spark_max("order_date").alias("max_date")
        ).collect()[0]
        
        print(f"   ë‚ ì§œ ë²”ìœ„: {date_stats['min_date']} ~ {date_stats['max_date']}")
        
        # ì¶”ê°€ í†µê³„
        product_count = df.select("product_name").distinct().count()
        total_amount = df.agg({"amount": "sum"}).collect()[0][0]
        
        print(f"   ìƒí’ˆ ì¢…ë¥˜: {product_count}ê°œ")
        print(f"   ì´ ì£¼ë¬¸ ê¸ˆì•¡: ${total_amount:,.2f}")
    
    print()
```

```python
# 4. íŒŒí‹°ì…”ë‹ ë¶„ì„
print("ğŸ—‚ï¸ íŒŒí‹°ì…”ë‹ ë¶„ì„:")

# íŒŒí‹°ì…˜ë³„ íŒŒì¼ ë¶„í¬
partition_dirs = glob.glob(f"{table_path}/data/order_date=*")
print(f"íŒŒí‹°ì…˜ ê°œìˆ˜: {len(partition_dirs)}")

for partition_dir in sorted(partition_dirs):
    partition_name = os.path.basename(partition_dir)
    files_in_partition = glob.glob(f"{partition_dir}/*.parquet")
    
    total_size = sum(os.path.getsize(f) for f in files_in_partition)
    total_records = 0
    
    for file_path in files_in_partition:
        df = spark.read.parquet(file_path)
        total_records += df.count()
    
    print(f"ğŸ“ {partition_name}:")
    print(f"   íŒŒì¼ ê°œìˆ˜: {len(files_in_partition)}")
    print(f"   ì´ í¬ê¸°: {total_size:,} bytes")
    print(f"   ì´ ë ˆì½”ë“œ: {total_records:,}ê°œ")
    print()
```

```python
# 5. íŒŒì¼ ë‚´ìš© ìƒ˜í”Œë§
print("ğŸ“‹ ê° ë°ì´í„° íŒŒì¼ì˜ ì‹¤ì œ ë‚´ìš© ìƒ˜í”Œ:")

for i, file_path in enumerate(data_files[:3]):
    print(f"\nğŸ“„ íŒŒì¼ #{i+1} ìƒ˜í”Œ ë°ì´í„°:")
    df = spark.read.parquet(file_path)
    df.show(5, truncate=False)
    
    print("ğŸ“Š ì»¬ëŸ¼ë³„ í†µê³„:")
    df.describe().show()
```

```python
# 6. Delete Files íƒìƒ‰ (MOR í…Œì´ë¸”ì¸ ê²½ìš°)
print("ğŸ—‘ï¸ Delete Files íƒìƒ‰:")

delete_files = glob.glob(f"{table_path}/**/*delete*.parquet", recursive=True)
if delete_files:
    print(f"ë°œê²¬ëœ Delete íŒŒì¼: {len(delete_files)}ê°œ")
    for delete_file in delete_files:
        print(f"  ğŸ“„ {delete_file}")
else:
    print("âŒ Delete íŒŒì¼ ì—†ìŒ (COW ë°©ì‹ì´ê±°ë‚˜ ì‚­ì œ ì‘ì—…ì´ ì—†ì—ˆìŒ)")

# ì‚­ì œ ì‘ì—… ì‹œë®¬ë ˆì´ì…˜
print("\nğŸ§ª ì‚­ì œ ì‘ì—… ì‹¤í—˜:")
print("ì¼ë¶€ ë°ì´í„°ë¥¼ ì‚­ì œí•˜ì—¬ Delete Files ìƒì„±í•´ë³´ê¸°...")

# íŠ¹ì • ì¡°ê±´ìœ¼ë¡œ ì‚­ì œ
spark.sql("DELETE FROM demo.blog.orders WHERE order_id <= 5")

# ì‚­ì œ í›„ ë‹¤ì‹œ í™•ì¸
delete_files_after = glob.glob(f"{table_path}/**/*delete*.parquet", recursive=True)
print(f"ì‚­ì œ í›„ Delete íŒŒì¼: {len(delete_files_after)}ê°œ")
```

```python
# 7. Data Layer ìš”ì•½
print("ğŸ¯ Data Layer ë¶„ì„ ê²°ê³¼ ìš”ì•½:")
print("=" * 50)

total_data_files = len(data_files)
total_size = sum(os.path.getsize(f) for f in data_files)
total_records = spark.sql("SELECT COUNT(*) FROM demo.blog.orders").collect()[0][0]

print(f"ğŸ“Š ì´ ë°ì´í„° íŒŒì¼ ìˆ˜: {total_data_files}")
print(f"ğŸ’¾ ì´ ë°ì´í„° í¬ê¸°: {total_size:,} bytes ({total_size/1024/1024:.2f} MB)")
print(f"ğŸ“‹ ì´ ë ˆì½”ë“œ ìˆ˜: {total_records:,}")
print(f"ğŸ“ íŒŒí‹°ì…˜ ìˆ˜: {len(partition_dirs)}")
print(f"ğŸ“„ í‰ê·  íŒŒì¼ í¬ê¸°: {total_size/total_data_files:,.0f} bytes")
print(f"ğŸ“Š íŒŒì¼ë‹¹ í‰ê·  ë ˆì½”ë“œ: {total_records/total_data_files:.0f}ê°œ")

print("\nâœ… Data Layer ë¶„ì„ ì™„ë£Œ!")
```
